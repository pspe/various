```python
%matplotlib notebook

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from itertools import product, combinations


import pandas as pd
import csv
import sys


def unicode_csv_reader (utf8_data, dialect=csv.excel, **kwargs):
    csv_reader = csv.reader (utf8_data, dialect=dialect, **kwargs)
    for row in csv_reader:
        yield [unicode(cell, 'utf-8') for cell in row]


        
data_path = "./"
print ("arguments: "+str (sys.argv));

filename = data_path+"projection.csv"
#filename = data_path+"testNames.csv"

# Load the data
#texts  = pd.read_csv (data_path+"corporateNames.csv")

reader = unicode_csv_reader (open (filename))
data = [txt[0] for txt in reader if len(txt) > 0]

return data

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

#draw cube
r = [-1, 1]
for s, e in combinations(np.array(list(product(r,r,r))), 2):
    if np.sum(np.abs(s-e)) == r[1]-r[0]:
        ax.plot3D(*zip(s,e), color="b")


plt.show()
```

```python

import pandas as pd
import csv
import sys
from math import *
import operator
import numpy as np

data_path = "./"

filename = data_path+"projection.csv"


def loadCSV (filename):
    data = []
    f = open(filename, 'rt')
    try:
        reader = csv.reader(f)
        for row in reader:
            data.append (row)
    finally:
        f.close()
    return data
        

def extractLine (line):
    dct = {"Image" : line[0],
           "Index" : int(line[1]),
           "pDet" : float(line[2]),
           "pSub" : float(line[3]),
           "p" : [float(x) for x in line[4:13]],
           "psx" : float(line[13]),
           "psy" : float(line[14]),
           "lDet" : float(line[15]),
           "lSub" : float(line[16]),
           "l" : [float(x) for x in line[17:26]],
           "lsx" : float(line[26]),
           "lsy" : float(line[27])
          }
    return dct

def length (vec):
    return sqrt (sum ([u**2 for u in vec]))
    
def unit (vec):
    return [c/length (vec) for c in vec]
    
def prod (v0, v1):
    return reduce (operator.add, map (operator.mul, v0, v1))
    
def diff (v0, v1):
    return [c - k for c,k in zip (v0, v1)]
    
def distortion (coeff):
#    x = unit (coeff[0:3])
#    y = unit (coeff[3:6])
#    z = unit (coeff[6:9])
    x = unit (coeff[0:2]+[0])
    y = unit (coeff[3:5]+[0])
    z = unit (coeff[6:9])

    return prod (diff (x, y), diff (z, y))/2.0
    
    
print distortion ([1,0,0,0,1,0,0,0,1])
print distortion ([0,1,0,-1,0,0,0,0,1])
print distortion ([0.5,0.5,-0.5,0.5,0,0,0,0,1])
print distortion ([1,0,0,0,1,0,0.1,0.1,1])
print distortion ([1,0,0.5,0,1,0.5,0,0,1])
print distortion ([0.5,0.5,-0.5,0.5,0,0,0.1,0.1,1])
for alpha in np.arange(0,2*pi,pi/16.0):   
    print "alpha = ",360*alpha/(2*pi)," --> ",distortion ([cos(alpha),-sin(alpha),sin(alpha),cos(alpha),0,0,0.0,0.0,1])
    
d = loadCSV (filename)
line = d[2]
x = {1 : line[4:13], 2 : line[13]}
#print "d[99:100] = ",d[98:100]
d = d[2:99]
data = [extractLine (x) for x in d]

#print distortion(data[0]["p"])
result = [(distortion (x["p"]), distortion(x["l"])) for x in data]
print result[3][0] > result[3][1]
classification = ['L' if x[0]>x[1] else 'P' for x in result]
#test = [x[1] for x in result]
#print test

imageName = [x['Image'] for x in data]
classification = zip (imageName, classification, result)
for l in classification:
    print l[0],' --> ',l[1],'     ',l[2]
    
```
